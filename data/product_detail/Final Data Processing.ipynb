```python
import numpy as np
import pandas as pd
import json
import os
```

```python
# DATA Path
DATAPATH_JSON = './json'
```

- Product info stored in product_rows and transfered into pandas table
- Product overview is stored in a dict

```python
product_rows = []
product_overview = {}
for filename in os.listdir(DATAPATH_JSON):
    with open(f'{DATAPATH_JSON}/{filename}', 'r') as f:
        product = json.load(f)
        id_, overview = product['id'], product['overview']
        product_overview[id_] = overview
        # remove overview from the product dict
        product.pop('overview')
        # title is the same as id
        product.pop('title')
        # store product in a pandas dataframe
        product_rows.append(product)        
```

```python
df_1 = pd.DataFrame(product_rows)
df_1
```
